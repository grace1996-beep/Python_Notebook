```python
import pandas as pd
```

loading datasets and doing cleaning before merging

```python
#first loading my celll_geo csv file
df=pd.read_csv("cells_geo.csv",delimiter=",")
df.rename(columns={"CDELL_ID":"CELL_ID"},inplace=True)
df.rename(columns={"VILLES":"CITY"},inplace=True)
df.rename(columns={"SITE_CODE":"SITE_ID"},inplace=True)
df
```

```python
#droping duplicates
df1=df.drop_duplicates()
#deleting rows with no data at all
dfna=df1.dropna(how='all')
#delete missing values in dataframe
df_final=dfna.dropna()
df_final
```

```python
#loading the telcom datasets and renamimg some of the columns ie cleaning
df1=pd.read_csv("Telcom_dataset.csv")
#to rename the columns PRODUTC TO PRODUCT
df1.rename(columns={"PRODUTC":"PRODUCT"},inplace=True)
df1.rename(columns={"DATETIME":"DATE_TIME"},inplace=True)
#to cleandf1
#to drop duplicates
dfrr=df1.drop_duplicates()
#to remove missing values
dfr3=dfrr.dropna()
#deleterows without data
dfall1=dfr3.dropna(how='all')
df2=pd.read_csv("Telcom_dataset2.csv")
#to clean to drop duplicates
dfm=df2.drop_duplicates()
#to remove mising values
dd=dfm.dropna()
#to remove rows without data
dfall2=dd.dropna(how='all')
df2.rename(columns={"DW_A_NUMBER":"DW_A_NUMBER_INT"},inplace=True)
df2.rename(columns={"DW_B_NUMBER":"DW_B_NUMBER_INT"},inplace=True)
df3=pd.read_csv("Telcom_dataset3.csv")

df3.rename(columns={"CELLID":"CELL_ID"},inplace=True)
df3.rename(columns={"SIET_ID":"SITE_ID"},inplace=True)
#to remove duplicates
dl=df3.drop_duplicates()
#to remove missing values
dl2=dl.dropna()
#to remove empty rows
dfall3=dl2.dropna(how='all')
dfs=pd.concat((dfall1,dfall2,dfall3),axis=0,ignore_index=True)
dfs


```
